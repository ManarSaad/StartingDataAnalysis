```python
# DATA MANIPULATION WITH PANDAS

import pandas as pd
df=pd.read_csv('Films.csv',delimiter=';')
df.head()
```

```python
# add a column of movies hours
df['Hours']=[2,1,2,1,3]
df.head()
```

```python
# leet's select a movie that is awarded
df[df.Awards=='Yes']
```

```python
# let's add a column that have information about the year and name of the movie 
combine_lambda = lambda row: \
    '{} - {}'.format(row.Title,
                     row.Year)
df['Information']=df.apply(combine_lambda, axis=1)
```

```python
df.head()
```

```python
# now, i want to change the name of -Title- to -Name-

df.rename(columns={
    'Title': 'Name'},
    inplace=True)
df.head()
```

```python
#what if i want to add a column to which is True if the movie not awarded and the year is greater than or equal 1990
# and False otherwise 
df['Not Awarded']=df.apply(lambda x: True if (x.Year >= 1990) & (x.Awards=='No')
else False, 
axis=1)
df.head()
```
